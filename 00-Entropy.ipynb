```python
import numpy as np

from plotly.offline import init_notebook_mode, iplot
import plotly.graph_objs as go

init_notebook_mode(connected=True)
```

```python
def entropy(xs):
    return (-xs*np.log(xs)).sum()

entropy(np.array([.5, .5]))
```

```python
entropy(np.array([0, 1]))
```

```python
p = np.linspace(0.01, 0.99, 99)
q = 1 - p

pq = np.array([p, q]).T
```

```python
p
```

```python
es = (-pq*np.log(pq)).sum(axis=1)
```

```python
iplot([go.Scatter(x=p, y=es)])
```
